```r
cat(sprintf("Children mu=%.3f sdv=%.3f\n", mean(galton$child), sd(galton$child)))
cat("\n")
cat(sprintf("Fathers mu=%.3f sdv=%.3f\n", mean(galton$parent), sd(galton$parent)))
```

```r
?t.test
```

```r
t.test(galton$child, galton$parent, paired=T)
```

```r
par(mfrow=c(1,2))
boxplot(galton$child, xlab="children", ylab="density(%)", main="", col="lightblue")
boxplot(galton$parent, xlab="parent", ylab="density(%)", main="", col="salmon")
```

```r
# par(mfrow=c(1,1), mar=c(bottom=2.3,left=2,top=2,right=1))

#--- simple plot
boxplot(galton$parent, galton$child,
        names=c( "parents", "children"),
        main="Galton data", 
        xlab="children x parents", ylab="heigth",
        col = c("lightblue", "salmon"))
```

### Boxplot

```r
boxplot(child~parent, data=galton)
```

```r
rainbow(n=11)
```

```r
boxplot(child~parent, data=galton, col=rainbow(n=11))
```

```r
par(mfrow=c(1,1))
plot(galton$child,galton$parent, main="Galton data", xlim=c(60,80), ylim=c(60,80))
```

```r
lm(child~parent, data=galton)
```

```r
fit = lm(child~parent, data=galton)
```

```r
round(coefficients(fit), 4)
```

## Linear plot: y = a*x + b

**abline()**

```r
par(mfrow=c(1,1))
plot(galton$child,galton$parent, main="Galton data", xlim=c(60,80), ylim=c(60,80), col="navy")
abline(coefficients(fit), col="red")
```

### Correlation

```r
round(cor(galton$child,galton$parent), 4)
```
