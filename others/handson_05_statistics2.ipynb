### Binomial distribution - 1 coin

[wiki - Binomial distribution](https://en.wikipedia.org/wiki/Binomial_distribution)

tossin coins  
**random binomial**  
rbinom(n, size, prob)  

**fair coin** p=.5
**biased coin** p=.3  

size = 1 - head and tail

```r
?rbinom
?hist
```

```r
n = 1000
size = 1
prob =.5

sbinoms = rbinom(n, size, prob)

hist(sbinoms, main="Simulating binomial distribution / 1 coin", col = "cyan")
```

```r
hist(sbinoms, main="Simulating binomial distribution / 1 coin", col = "cyan", freq=F)
```

### Binomial distribution - 2 coins

size = 2

```r
n = 1000
size = 2 # 2 coins
prob =.5

sbinoms = rbinom(n, size, prob)

hist(sbinoms, main="Simulating binomial distribution / 1 coin", col = "cyan")
```

```r
### ggplot package

library(ggplot2)
```

```r
df = data.frame(samps=sbinoms)

ggplot(data=df, aes(samps)) +
  geom_histogram(binwidth=.5, colour="navy", fill="cyan", alpha=.3) +
  labs(title="Binomial frequency plot - 2 fair coins", y="#")
```

```r
ggplot(data=df, aes(samps)) +
  geom_histogram(aes(y=..density..), binwidth=.5, colour="navy", fill="cyan", alpha=.3) +
  labs(title="Binomial distribution - 2 fair coins", y="percentual (%)")
```

### Normal distribution

[wiki - Normal distribution](https://en.wikipedia.org/wiki/Normal_distribution)

parameters:
  * mu = mean  
  * std = standar deviation
  

?rnorm  
  
rnorm(n, mean = 0, sd = 1)

```r
n = 1000

mu = 1
std = 1
```

```r
samps = rnorm(n, mu, std)
df = data.frame(samps)

ggplot(data=df, aes(samps)) +
  geom_histogram(binwidth=.5, colour="navy", fill="cyan", alpha=.3)
```

```r
ggplot(data=df, aes(samps)) +
  geom_histogram(aes(y=..density..), binwidth=.5, colour="navy", fill="cyan", alpha=.3)
```

```r
ggplot(data=df, aes(samps)) +
  geom_histogram(aes(y=..density..), binwidth=.5, colour="navy", fill="cyan", alpha=.3) +
  stat_function(fun=dnorm, args=list(mean=mu, sd=std), col="blue")+
  labs(title="Normal distribution", y="percentual (%)")
```

```r
samps1 = rnorm(n, mean=0, sd=1)
samps2 = rnorm(n, mean=2, sd=1)
samps3 = rnorm(n, mean=4, sd=3)

df = data.frame(samps1, samps2, samps3)
head(df)
```

```r
ggplot(data=df) +
  geom_histogram(aes(x=samps1, y=..density..), binwidth=.5, colour="blue", fill="cyan", alpha=.3) +
  geom_histogram(aes(x=samps2, y=..density..), binwidth=.5, colour="blue", fill="tomato", alpha=.3) +
  geom_histogram(aes(x=samps3, y=..density..), binwidth=.5, colour="blue", fill="green", alpha=.3) +
  stat_function(fun=dnorm, args=list(mean=0, sd=1), col="navy")+
  stat_function(fun=dnorm, args=list(mean=2, sd=1), col="red")+
  stat_function(fun=dnorm, args=list(mean=4, sd=3), col="darkgreen")+
  labs(title="Normal distributions", y="percentual (%)")
```

### Poisson distribution

[wiki - Poisson distribution](https://en.wikipedia.org/wiki/Poisson_distribution)

rare events: 
**random poisson**  
rpois(n, lambda)

lambda = 0.1

```r
?rpois
```

```r
n = 1000
lambda = 1

samps = rpois(n, lambda)
df = data.frame(samps)

ggplot(data=df, aes(samps)) + # xlim(0,10) + ylim(0,0.35) +
  geom_histogram(aes(y=..density..), binwidth=.5, colour="navy", fill="cyan", alpha=.3) +
  labs(title="Poisson distribution", y="percentual (%)", x="k") 
```

```r
lambda = 1

samps1 = rpois(n, lambda=1)
samps2 = rpois(n, lambda=5)
samps3 = rpois(n, lambda=10)
df = data.frame(samps1, samps2, samps3)
head(df)
```

```r
ggplot(data=df) + # xlim(0,50)+
  geom_histogram(aes(x=samps1, y=..density..), binwidth=.5, colour="navy", fill="cyan",    alpha=.3) +
  geom_histogram(aes(x=samps2, y=..density..), binwidth=.5, colour="navy", fill="tomato", alpha=.3) +
  geom_histogram(aes(x=samps3, y=..density..), binwidth=.5, colour="navy", fill="green",   alpha=.3) +
  labs(title="Poisson distributions", y="percentual (%)", x="k") 
```
